# np.random.seed(seed)に注意

```python
from tdgl import tdgl
from multiprocessing import Pool
import multiprocessing
import numpy as np
import time
import matplotlib.pyplot as plt

def wrapper(args):
    return tdgl(*args,save=True)

def multi_process(argList):
    n_cores = multiprocessing.cpu_count()
    p = Pool( n_cores )
    output = p.map(wrapper, argList)
    p.close()
```

# random array not seed values

```python
save_array = np.arange(0,1e-2+1e-4,1e-4)
hv_array = np.linspace(-4,-2,101)
random_array = np.arange(2,4,1)
```

```python
arg_list = []
for hv,sv in zip(hv_array,save_array):
    for random_seed in random_array:
        tmp_arg = (hv,round(sv,5),random_seed)
        arg_list.append(tmp_arg)
```

```python
tmp_arg
```

```python
block = 8
iteration = int( -( -len(arg_list)//block ) )
```

```python
now = time.time()
for i in range(iteration):
    part_arg = arg_list[block*i:block*(i+1)]
    multi_process(part_arg)
print("time:",time.time() - now)
```
